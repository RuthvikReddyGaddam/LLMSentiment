```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from win32con import MOUSEEVENTF_VIRTUALDESK
```

```python
# Download latest version
path = kagglehub.dataset_download("utkarshx27/movies-dataset")
df = pd.read_csv(path+"/movie_dataset.csv", index_col=0)
df

```

```python
axes = plt.axes()
sns.heatmap(df.isna().transpose(), cbar=False, ax=axes)
plt.xlabel("Columns")
plt.ylabel("missing values")
plt.show()
```

```python
dfFinal = df.copy()
dfFinal = dfFinal.drop(columns=["homepage"])
removeFInal = dfFinal.copy()
removeMissing = dfFinal.dropna()
removeMissing
```

```python
removeMissing.genres.value_counts().reset_index().sort_values(by="count", ascending=False)
```

```python
removeMissing = removeMissing.copy()
removeMissing['words in overview'] = removeMissing['overview'].str.split().str.len()
removeMissing
```

```python
removeMissing.loc[removeMissing['words in overview'].between(15,50), "overview"]
```

```python
movieMissing = removeMissing.copy()
movieMissing = movieMissing.loc[movieMissing['words in overview'] >= 15]
movieMissing
```

```python
import ast
movieMissing['production_companies'] = movieMissing['production_companies'].apply(lambda x: [d['name'] for d in ast.literal_eval(x)] if isinstance(x, str) else None)
movieMissing['production_countries'] = movieMissing['production_countries'].apply(lambda x: [d['name'] for d in ast.literal_eval(x)] if isinstance(x, str) else None)
movieMissing['spoken_languages'] = movieMissing['spoken_languages'].apply(lambda x: [d['name'] for d in ast.literal_eval(x)] if isinstance(x, str) else None)
movieMissing['crew'] = movieMissing['crew'].apply(lambda x: [f"{d['name']} {d['job']}" for d in ast.literal_eval(x)] if isinstance(x, str) else None)
movieMissing
```

```python
movieMissing['spoken_languages'].value_counts()
```

```python
movieMissing['production_countries'].value_counts()
```

```python
movieMissing['production_companies'].value_counts()
```

```python
movieMissing.drop(["words in overview"], axis=1, inplace=True)
movieMissing["taggedDescription"] = movieMissing['id'].astype(str) + " " + movieMissing['overview'].astype(str)
```
